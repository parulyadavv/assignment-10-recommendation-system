```python
import pandas as pd
import numpy as np
```

```python
book = pd.read_csv('book.csv')
book
```

```python
book.info()
```

```python
#renaming the columns
book1 = book.rename({'User.ID':'userid','Book.Title':'title','Book.Rating':'rating'}, axis= 1)
```

```python
book1
```

```python
#number of unique users in the dataset
len(book1.userid.unique())
```

```python
len(book1.title.unique())
```

```python
rating_matrix = book1.pivot_table(index='userid',columns='title',values='rating').reset_index(drop=True)
rating_matrix
```

```python
rating_matrix.index = book1['userid'].unique()
```

```python
rating_matrix
```

```python
#imputing nans with 0
rating_matrix.fillna(0,inplace= True)
```

```python
rating_matrix
```

```python
#calculating cosine similarity between users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1-pairwise_distances(rating_matrix.values, metric = 'cosine')
```

```python
user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)
```

```python
#set the column and index name to userids
user_sim_df.index = book1.userid.unique()
user_sim_df.columns = book1.userid.unique()
```

```python
user_sim_df
```

```python
# Nullifying diagonal values
np.fill_diagonal(user_sim,0)
user_sim_df
```

```python
#most similar users
user_sim_df.idxmax(axis=1)
```
